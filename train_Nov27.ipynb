```python
from sklearn.decomposition import PCA
```

```python
X_pro = pro_df.values
```

```python
pca = PCA(n_components=2)
pca.fit(X_pro)
```

```python
pca.components_
```

```python
pca.explained_variance_
```

```python
X_pro_pca = pca.transform(X_pro)
```

```python
X_pro_pca.shape
```

```python
# for two component
plt.scatter(x=X_pro_pca[:,0], y=X_pro_pca[:,1])
```

```python
plt.scatter(x=X_pro_pca[:,0], y=X_pro_pca[:,1], c=sum_df['Proteomics'])
```

```python
plt.scatter(x=X_pro_pca[:,0], y=X_pro_pca[:,1], c=sum_df['RNAseq'])
```

```python
plt.scatter(x=X_pro_pca[:,0], y=X_pro_pca[:,1], c=sum_df['Clinical'])
```

```python
sum_df['mislabeled'] = np.where(sum_df['RNAseq'] != sum_df['Proteomics'], 1, 0)
plt.scatter(x=X_pro_pca[:,0], y=X_pro_pca[:,1], c=sum_df['mislabeled'])
```
